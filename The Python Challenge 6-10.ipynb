# Challenge 10
http://www.pythonchallenge.com/pc/return/bull.html

## Solution http://www.pythonchallenge.com/pc/return/5808.html

The URL takes us to an image of a ...wait for it...bull!  Clicking on it, we see a sequence a = [1, 11, 21, 1211, 111221,

This is not just any sequence, but Conway's look-and-say-sequence.  We are to calcualte len(a[30]).

```python
a = ['1'] #seed with 1, but really could start anywhere 
nxt_ele =''

for n in range(30): #find the first 30 elements of the sequence
    k=0
    j=0
    
    while k < len(a[n]): #n iterates through elements of array of squences
        while j < len(a[n]) and a[n][k]==a[n][j]: #k, j iterates thorugh digits in each element
            j+=1 #j counts apperances of digit at index k
      
        nxt_ele += str(j-k) #how many times was this digit seen?
                            #j > k on each iteration, so must subtract: j-k to get correct #seen 
        nxt_ele += str(a[n][k]) #copy the looked at digit after saying how many times it was seen
        k=j #start count at next new element, but j has not been reset to 0  
            
    a.append(nxt_ele)
    nxt_ele ='' #reset element
```
